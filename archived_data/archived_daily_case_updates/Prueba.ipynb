```python
from os import scandir, getcwd
import pandas as pd
import numpy as np

def ls(ruta = getcwd()):
    return [arch.name for arch in scandir(ruta) if arch.is_file()]
```

```python
def soloCSV():
    salida = []
    for i in ls():
        largo = len(i)
        if("csv" == i[ largo - 3 : largo]):
            salida.append(i)
            print(i)
    return salida

def DataUnida():
    salida = []
    for i in soloCSV():
        data_aux = pd.read_csv(i)
        print(len(data_aux))
        salida.append(data_aux)
    return pd.concat(salida)
```
