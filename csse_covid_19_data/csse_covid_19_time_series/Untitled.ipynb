```python
import pandas as pd
```

```python
data = pd.read_csv("time_series_covid19_confirmed_global.csv")
```

```python
data.head()
```

```python
data_salida = data[["Province/State","Country/Region","Lat","Long"]]
```

```python
columna_anterior = ""
for columna in data.columns[5:]:
    if(columna_anterior == ""):
        #print("no se hace nada")
        pass
    else:
        #print(columna + " y " + columna_anterior)
        data_salida[columna] = data[columna] - data[columna_anterior]
    columna_anterior = columna
```
