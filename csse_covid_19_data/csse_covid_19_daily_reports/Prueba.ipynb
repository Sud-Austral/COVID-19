```python
from os import scandir, getcwd
import pandas as pd
import numpy as np

def ls(ruta = getcwd()):
    return [arch.name for arch in scandir(ruta) if arch.is_file()]

def soloCSV():
    salida = []
    for i in ls():
        largo = len(i)
        if("csv" == i[ largo - 3 : largo]):
            salida.append(i)
            #print(i)
    return salida

def DataUnida():
    x = 1;
    salida = []
    for i in soloCSV():
        data_aux = pd.read_csv(i)
        print(str(x) + " " + str(len(data_aux)))
        salida.append(data_aux)
        x += 1
    return pd.concat(salida)
```

```python
data = DataUnida()
data.to_excel("Total.xlsx", index=False)
```
